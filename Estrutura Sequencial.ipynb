#### 8. Vamos criar um conversor de temperatura. Faça um Programa que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius.
$C = \frac{5}{9}(F-32)$

```python
# Solicita a temperatura em Fahrenheit ao usuário
Fahrenheit = float(input('Digite a Temperatura em Fahrenheit: '))

# Converte Fahrenheit para Celsius 
Celsius = (5/9) * (Fahrenheit - 32)

# Exibe a temperatura em Celsius
print(f'A temperatura em graus Celsius é {int(Celsius)}')
```

#### 9. Faça um Programa que peça a temperatura em graus Celsius, transforme e mostre em graus Fahrenheit.
$F = \frac{9}{5}C + 32$

```python
# Solicita a temperatura em Celsius ao usuário
Celsius = float(input('Digite a Temperatura em Celsius: '))

# Converte Celsius para Fahrenheit
Fahrenheit = (Celsius * 9/5) + 32

# Exibe a temperatura em Fahrenheit
print(f'A temperatura em Fahrenheit é {int(Fahrenheit)}')
```

#### 12. Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês.

```python
# Pergunta ao usuário quanto ele ganha por hora
Valor_Hora = float(input("Digite o valor das horas trabalhada: "))

# Pergunta ao usuário quantas horas ele trabalhou no mês
Horas_Trabalhadas = float(input("Quantas horas você trabalhou este mês? "))

# Calcula o salário bruto
Salario_Bruto =  Valor_Hora * Horas_Trabalhadas

# Exibindo o salário bruto
print(f"Seu salário bruto este mês é R$ {Salario_Bruto}")

```

#####  Calcule o salário bruto (horas * salario por hora)


```python
# Pergunta ao usuário quanto ele ganha por hora
ganho_por_hora = float(input("Quanto você ganha por hora? R$ "))

# Pergunta ao usuário quantas horas ele trabalhou no mês
horas_trabalhadas = float(input("Quantas horas você trabalhou este mês? "))

# Calcula o salário bruto
salario_bruto = ganho_por_hora * horas_trabalhadas

# Imprime o salário bruto
print("Seu salário bruto este mês é R$", salario_bruto)
```

##### Calcule o desconto do IR (11% do salário bruto)

```python
# Pergunta ao usuário quanto ele ganha por hora
ganho_por_hora = float(input("Quanto você ganha por hora? R$ "))

# Pergunta ao usuário quantas horas ele trabalhou no mês
horas_trabalhadas = float(input("Quantas horas você trabalhou este mês? "))

# Calcula o salário bruto
salario_bruto = ganho_por_hora * horas_trabalhadas

# Calcula o desconto do IR (11% do salário bruto)
desconto_ir = salario_bruto * 0.11

# Imprime o salário bruto e o desconto do IR
print("Seu salário bruto este mês é R$", salario_bruto)
print("O desconto do IR é R$", desconto_ir)
```

##### Calcule o desconto do INSS (8% do salário bruto)

```python
# Pergunta ao usuário quanto ele ganha por hora
ganho_por_hora = float(input("Quanto você ganha por hora? R$ "))

# Pergunta ao usuário quantas horas ele trabalhou no mês
horas_trabalhadas = float(input("Quantas horas você trabalhou este mês? "))

# Calcula o salário bruto
salario_bruto = ganho_por_hora * horas_trabalhadas

# Calcula o desconto do IR (11% do salário bruto)
desconto_ir = salario_bruto * 0.11

# Calcula o desconto do INSS (8% do salário bruto)
desconto_inss = salario_bruto * 0.08

# Imprime o salário bruto, o desconto do IR e o desconto do INSS
print("Seu salário bruto este mês é R$", salario_bruto)
print("O desconto do IR é R$", desconto_ir)
print("O desconto do INSS é R$", desconto_inss)
```
